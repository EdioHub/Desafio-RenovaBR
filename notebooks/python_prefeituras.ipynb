```python
# Remover linhas onde 'Resultado_Eleitoral' é NaN em df_2024
df_2024 = df_2024.dropna(subset=["Resultado_Eleitoral"])

# Verificar valores únicos em 'Resultado_Eleitoral' após remoção
print(df_2024["Resultado_Eleitoral"].unique())
```

```python
# Filtrar apenas candidatos eleitos para Prefeito ou Vice-prefeito
prefeitos_vice_eleitos = df_2024[
    (df_2024["Resultado_Eleitoral"] == "Eleito") &
    (df_2024["Cargo"].isin(["Prefeito", "Vice-prefeito"]))
]

# Contar quantos prefeitos e quantos vices eleitos existem por cidade
contagem = prefeitos_vice_eleitos.groupby(["Cidade", "Cargo"]).size().unstack(fill_value=0)

# Garantir que todas as cidades apareçam, mesmo que sem eleitos
todas_cidades = df_2024["Cidade"].unique()
contagem = contagem.reindex(todas_cidades, fill_value=0)

# Adicionar uma coluna indicando se a cidade tem 1 prefeito e 1 vice
contagem["Tem_Prefeito_e_Vice"] = (contagem.get("Prefeito", 0) == 1) & (contagem.get("Vice-prefeito", 0) == 1)

# Resetar índice para transformar 'Cidade' em coluna
contagem = contagem.reset_index()

# Exibir resultado
print(contagem[["Cidade", "Prefeito", "Vice-prefeito", "Tem_Prefeito_e_Vice"]])

```

```python
# Passo 1: Filtrar os candidatos que são prefeitos eleitos
prefeitos_eleitos = df_2024[
    (df_2024["Cargo"] == "Prefeito") &
    (df_2024["Resultado_Eleitoral"] == "Eleito")
]

# Passo 2: Contar quantos prefeitos eleitos há por cidade
contagem_prefeitos = prefeitos_eleitos.groupby("Cidade").size()

# Passo 3: Pegar apenas as cidades com exatamente 1 prefeito eleito
cidades_validas = contagem_prefeitos[contagem_prefeitos == 1].index.tolist()

# Passo 4: Manter todas as linhas (de qualquer cargo) dessas cidades válidas
df_prefeito_cidades = df_2024[df_2024["Cidade"].isin(cidades_validas)]

# (Opcional) Verificar o resultado
print("Cidades com exatamente 1 prefeito eleito:", cidades_validas)
print("Amostra dos dados filtrados:")
print(df_prefeito_cidades["Cidade"].value_counts())
```

```python
# Assumindo que seu dataframe 2020 se chama df_2020

# Filtrar apenas candidatos a Prefeito e Vice-prefeito
df_2020_filtered = df_2020[df_2020['Cargo'].isin(['Prefeito', 'Vice-prefeito'])]

# Função para calcular porcentagem por grupo
def calcular_percentual(df, group_cols):
    contagem = df.groupby(group_cols).size().reset_index(name='contagem')
    total = df.groupby('Cargo').size().reset_index(name='total')
    df_perc = contagem.merge(total, on='Cargo')
    df_perc['percentual'] = (df_perc['contagem'] / df_perc['total']) * 100
    return df_perc

# 1. Por gênero
df_genero = calcular_percentual(df_2020_filtered, ['Cargo', 'Genero'])

# 2. Por raça
df_raca = calcular_percentual(df_2020_filtered, ['Cargo', 'Raca'])

# 3. Por gênero e raça
df_genero_raca = calcular_percentual(df_2020_filtered, ['Cargo', 'Genero', 'Raca'])

# --- Gráficos ---

sns.set(style="whitegrid")

# Gráfico: Prefeito e Vice por Gênero
plt.figure(figsize=(8,5))
sns.barplot(data=df_genero, x='Genero', y='percentual', hue='Cargo')
plt.title('Percentual de Candidatos por Cargo e Gênero (2020)')
plt.ylabel('Percentual (%)')
plt.xlabel('Gênero')
plt.legend(title='Cargo')
plt.savefig("../figuras/pref_cargo_genero20.png", dpi=300)
plt.show()


# Gráfico: Prefeito e Vice por Raça
plt.figure(figsize=(10,5))
sns.barplot(data=df_raca, x='Raca', y='percentual', hue='Cargo')
plt.title('Percentual de Candidatos por Cargo e Raça (2020)')
plt.ylabel('Percentual (%)')
plt.xlabel('Raça')
plt.legend(title='Cargo')
plt.savefig("../figuras/pref_cargo_raça20.png", dpi=300)
plt.show()

# Gráfico: Prefeito e Vice por Gênero e Raça (agrupado)
plt.figure(figsize=(14,6))
sns.barplot(data=df_genero_raca, x='Raca', y='percentual', hue='Genero', palette='pastel', ci=None)
plt.title('Percentual de Candidatos por Raça e Gênero (2020)')
plt.ylabel('Percentual (%)')
plt.xlabel('Raça')
plt.legend(title='Gênero')
plt.savefig("../figuras/pref_cargo_genero_raça20.png", dpi=300)
plt.show()

# --- Tabelas (Exibição simplificada) ---

print("Tabela percentual por Cargo e Gênero:")
print(df_genero.pivot(index='Genero', columns='Cargo', values='percentual').round(2))

print("\nTabela percentual por Cargo e Raça:")
print(df_raca.pivot(index='Raca', columns='Cargo', values='percentual').round(2))

print("\nTabela percentual por Cargo, Gênero e Raça:")
print(df_genero_raca.pivot_table(index=['Raca', 'Genero'], columns='Cargo', values='percentual').round(2))
```

```python
# Filtrar dados válidos para Prefeito (2020)
df_receita_2020_prefeito = df_2020[
    (df_2020["Cargo"] == "Prefeito") &
    (df_2020["Valor_Receita"].notna()) &
    (df_2020["Valor_Receita"] > 0)
].copy()

# Padronizar texto de gênero
df_receita_2020_prefeito["Genero"] = df_receita_2020_prefeito["Genero"].replace({
    "MASC.": "MASCULINO",
    "FEM.": "FEMININO"
})

# Paleta de cores por gênero
cores_genero = {
    "MASCULINO": "#2196F3",
    "FEMININO": "#E91E63"
}

# Gráfico: Receita por Gênero (Prefeito - 2020)
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=df_receita_2020_prefeito,
    x="Genero",
    y="Valor_Receita",
    palette=cores_genero,
    showfliers=False
)
sns.stripplot(
    data=df_receita_2020_prefeito,
    x="Genero",
    y="Valor_Receita",
    color="black",
    alpha=0.3,
    jitter=True,
    size=2
)
plt.yscale("log")
plt.title("Distribuição de Receita por Gênero - Prefeito (2020)")
plt.xlabel("Gênero")
plt.ylabel("Valor da Receita (log)")
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('../figuras/pref_box_receita_genero_total.png', dpi=300)
plt.show()


# Tabela com estatísticas por gênero
df_gen_agg_2020 = df_receita_2020_prefeito.groupby("Genero").agg(
    receita_total=("Valor_Receita", "sum"),
    media_por_candidatura=("Valor_Receita", "mean"),
    desvio_padrao=("Valor_Receita", "std"),
    numero_candidaturas=("Valor_Receita", "count")
).reset_index()

# Cálculo do percentual da receita total
total_receita_2020 = df_gen_agg_2020["receita_total"].sum()
df_gen_agg_2020["percentual_receita"] = (df_gen_agg_2020["receita_total"] / total_receita_2020) * 100

# Reorganizando colunas e arredondando
df_gen_agg_2020 = df_gen_agg_2020[[
    "Genero",
    "numero_candidaturas",
    "receita_total",
    "media_por_candidatura",
    "desvio_padrao",
    "percentual_receita"
]].round(2)

# Exibição da tabela final
print("\nTabela - Receita de Campanha por Gênero (Prefeito - 2020):")
print(df_gen_agg_2020)
```

```python
# Filtrar dados válidos
df_receita_2020 = df_2020[
    (df_2020["Cargo"].isin(["Prefeito", "Vice-prefeito"])) &
    (df_2020["Valor_Receita"].notna()) &
    (df_2020["Valor_Receita"] > 0)
].copy()

# Padronizar texto
df_receita_2020["Genero"] = df_receita_2020["Genero"].replace({"MASC.": "MASCULINO", "FEM.": "FEMININO"})
df_receita_2020["Raca"] = df_receita_2020["Raca"].str.upper().str.strip()

# Paletas de cores
cores_raca = {
    "BRANCA": "#2196F3",
    "PARDA": "#FFC107",
    "PRETA": "#4CAF50",
    "AMARELA": "#9C27B0",
    "INDÍGENA": "#FF5722"
}
cores_genero = {
    "MASCULINO": "#2196F3",
    "FEMININO": "#E91E63"
}

# Coluna combinada de Raça e Gênero
df_receita_2020["Raca_Genero"] = df_receita_2020["Raca"] + " - " + df_receita_2020["Genero"]

# Ordem para exibição
ordem = [f"{r} - {g}" for r in cores_raca.keys() for g in ["FEMININO", "MASCULINO"]]

# Gráfico: Receita por Raça e Gênero (2020)
plt.figure(figsize=(16, 6))
sns.boxplot(
    data=df_receita_2020,
    x="Raca_Genero",
    y="Valor_Receita",
    palette=[cores_raca.get(rc.split(" - ")[0], "#cccccc") for rc in ordem],
    order=ordem,
    showfliers=False
)
sns.stripplot(
    data=df_receita_2020,
    x="Raca_Genero",
    y="Valor_Receita",
    color="black",
    order=ordem,
    alpha=0.3,
    jitter=True,
    size=2
)
plt.yscale("log")
plt.title("Distribuição de Receita por Raça e Gênero (2020)")
plt.xlabel("Raça e Gênero")
plt.ylabel("Valor da Receita (log)")
plt.xticks(rotation=45)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("../figuras/pref_box_receita_genero_total20.png", dpi=300)
plt.show()

# Tabela final: Receita por Raça e Gênero (2020)
df_raca_gen_agg_2020 = df_receita_2020.groupby("Raca_Genero").agg(
    numero_candidaturas=("Valor_Receita", "count"),
    receita_total=("Valor_Receita", "sum"),
    media_por_candidatura=("Valor_Receita", "mean"),
    desvio_padrao=("Valor_Receita", "std")
).reset_index()

# Cálculo do percentual da receita total
total_geral_2020 = df_raca_gen_agg_2020["receita_total"].sum()
df_raca_gen_agg_2020["percentual_receita"] = (df_raca_gen_agg_2020["receita_total"] / total_geral_2020) * 100

# Reorganizando colunas e arredondando
df_raca_gen_agg_2020 = df_raca_gen_agg_2020[[
    "Raca_Genero",
    "numero_candidaturas",
    "receita_total",
    "media_por_candidatura",
    "desvio_padrao",
    "percentual_receita"
]].round(2)

# Exibição
print("\nTabela - Receita de Campanha por Raça e Gênero (2020):")
print(df_raca_gen_agg_2020)
```

<span style="font-weight:bold; color:yellow; font-size:20px">Prefeitura 2024

<span style="font-weight:bold; color:yellow; font-size:15px">A distribuição das candidaturas à prefeitura em 2024 reforça a permanência de desigualdades estruturais no acesso aos espaços de poder. Do total de candidatos, 77,23% são homens e apenas 22,77% são mulheres para o cargo de prefeito. No caso dos vice-prefeitos, embora a participação feminina seja maior (46,53%), os homens ainda predominam com 53,47%. Esses dados evidenciam que, mesmo com avanços no debate sobre inclusão e equidade, as mulheres continuam sub-representadas, especialmente nos cargos majoritários de maior visibilidade e poder decisório.
<br>
No recorte racial, observa-se que a maioria dos candidatos se declara branca: 62,38% para o cargo de prefeito e 59,41% para o de vice-prefeito. Candidatos pardos representam 24,75% entre prefeitos e 23,76% entre vices, enquanto os pretos somam 11,88% e 16,83%, respectivamente. Candidatos amarelos aparecem apenas entre os prefeitos (0,99%), e não há registros de candidatos indígenas nesse recorte. Assim como na divisão por gênero, também se verifica uma expressiva concentração racial, em que pessoas brancas ocupam a maior parte dos espaços políticos, enquanto grupos racialmente minorizados continuam enfrentando barreiras significativas à participação eleitoral.
<br>
A combinação de raça e gênero revela com ainda mais nitidez os efeitos da exclusão interseccional. Homens brancos constituem o grupo mais numeroso entre os candidatos a prefeito (50,50%) e também têm forte presença como vices (32,67%). As mulheres brancas representam 11,88% das candidaturas à prefeitura e 26,73% à vice. Homens pardos seguem com 21,78% e 15,84%, enquanto as mulheres pardas são 2,97% e 7,92%, respectivamente. Já os homens pretos somam 3,96% das candidaturas a prefeito e 4,95% a vice, e as mulheres pretas 7,92% e 11,88%. Mulheres amarelas e indígenas não aparecem nesse recorte. Os dados deixam claro que, quanto mais marcadores de desigualdade se acumulam — como ser mulher e negra —, menor é a presença nos espaços de disputa política.
<br>
O contraste é ainda mais evidente ao se observar que mulheres pretas e pardas, apesar de representarem uma parte significativa da população brasileira, somam apenas 10,89% das candidaturas a prefeito e 19,8% a vice-prefeito. Isso demonstra que as iniciativas de inclusão atuais não têm sido suficientes para romper as barreiras estruturais que impedem a presença dessas mulheres na política em condições de igualdade.
<br>
Em síntese, a análise das candidaturas de 2024 confirma padrões históricos de exclusão na política brasileira: predominância de homens brancos, sub-representação feminina e baixa inclusão de pessoas negras. A desigualdade de gênero e raça continua sendo um dos principais desafios para a efetivação de uma democracia verdadeiramente representativa.
