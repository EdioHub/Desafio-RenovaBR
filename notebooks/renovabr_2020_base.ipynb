```r
# 1. Filtrar apenas Prefeito e Vice-Prefeito eleitos
prefeitos_vice_eleitos <- candidatos_2020 %>%
  filter(Cargo %in% c("Prefeito", "Vice-prefeito") & Resultado_Eleitoral == "Eleito")

# Contar separadamente o total de prefeitos e vices eleitos
total_prefeitos <- prefeitos_vice_eleitos %>%
  filter(Cargo == "Prefeito") %>%
  nrow()

total_vices <- prefeitos_vice_eleitos %>%
  filter(Cargo == "Vice-prefeito") %>%
  nrow()

cat("Total de prefeitos eleitos:", total_prefeitos, "\n")
cat("Total de vice-prefeitos eleitos:", total_vices, "\n")

# 2. Contar número de eleitos por cidade
resumo_por_cidade <- prefeitos_vice_eleitos %>%
  group_by(Cidade) %>%
  summarise(
    n_prefeito = sum(Cargo == "Prefeito"),
    n_vice = sum(Cargo == "Vice-prefeito"),
    .groups = "drop"
  )

# 3. Filtrar cidades onde falta prefeito OU vice
cidades_incompletas <- resumo_por_cidade %>%
  filter(n_prefeito == 0 | n_vice == 0)

# 4. Exibir as cidades que devem ser retiradas
cat("Número de cidades com ausência de prefeito ou vice:", nrow(cidades_incompletas), "\n")
print(cidades_incompletas)
```

 <span style="font-weight:bold; color:yellow; font-size:20px"> Devido ao fato de não termos um prefeito eleito nos dados do Rio de Janeiro, vamos tratar essa situação como uma inconsistência no banco de dados. Assim, retiraremos as candidaturas de prefeitos e vice-prefeitos dessa cidade do nosso banco de dados candidatos_2020.</span>

```r
# Remover candidatos a Prefeito ou Vice-prefeito das cidades incompletas
candidatos_2020 <- candidatos_2020 %>%
  filter(!(Cargo %in% c("Prefeito", "Vice-prefeito") & Cidade %in% cidades_incompletas$Cidade))
```

```r
# Análise de dados de candidaturas validadas - Gênero e Raça

# 1. Análise de Gênero - Gráfico de Pizza
contagem_genero <- candidaturas_validadas %>%
  count(Genero) %>%
  mutate(porcentagem = n / sum(n) * 100)

# Gráfico de pizza com cores personalizadas para gênero
pie_genero <- ggplot(contagem_genero, aes(x = "", y = n, fill = Genero)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y", start = 0) +
  geom_text(aes(label = paste0(round(porcentagem), "%")), 
            position = position_stack(vjust = 0.5)) +
  scale_fill_manual(values = c("FEM." = "deeppink", "MASC." = "steelblue")) +
  labs(title = "Distribuição por Gênero para candidatos a vereador 2020") +
  theme_void() +
  theme(plot.title = element_text(hjust = 0.5))

print(pie_genero)

# 2. Análise de Raça - Gráfico de Barras Verticais
contagem_raca <- candidaturas_validadas %>%
  count(Raca) %>%
  arrange(desc(n)) %>%
  mutate(Raca = factor(Raca, levels = Raca))

library(ggplot2)

bar_raca <- ggplot(contagem_raca, aes(x = Raca, y = n, fill = Raca)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = n), vjust = -0.3, size = 3.5) +
  labs(title = "Distribuição por Raça de candidatos a vereador 2020", 
       x = "Raça", 
       y = "Quantidade") +
  scale_fill_manual(values = c(
    "PRETA" = "#4F4F4F",         # Cinza meio escuro
    "BRANCA" = "#A9C0D8",        # Azul acinzentado claro
    "AMARELA" = "#FFD700",       # Amarelo
    "PARDA" = "#8B4513",         # Marrom
    "INDÍGENA" = "#F08080"       # Vermelho claro
  )) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        plot.title = element_text(hjust = 0.5))

print(bar_raca)
```

```r
library(ggplot2)
library(dplyr)
library(tidyr)

media_receita_corrigida <- candidatos_eleitos %>%
  filter(Resultado_Eleitoral == "Eleito", Genero %in% c("MASC.", "FEM.")) %>%
  group_by(Partido, Genero) %>%
  summarise(
    Media_Receita = mean(Valor_Receita, na.rm = TRUE),
    Qtd = n(),
    .groups = "drop"
  ) %>%
  mutate(Media_Receita = round(Media_Receita, 2)) %>%
  pivot_wider(
    names_from = Genero,
    values_from = c(Media_Receita, Qtd),
    names_sep = "_"
  ) %>%
  rename(
    Media_FEM. = Media_Receita_FEM.,
    Media_MASC. = Media_Receita_MASC.
  ) %>%
  mutate(
    Proporcao_Masc_Fem = case_when(
      !is.na(Qtd_MASC.) & !is.na(Qtd_FEM.) & Qtd_MASC. > Qtd_FEM. ~ paste0("MASC ", round(Qtd_MASC. / Qtd_FEM., 1), ":1"),
      !is.na(Qtd_MASC.) & !is.na(Qtd_FEM.) & Qtd_FEM. > Qtd_MASC. ~ paste0("FEM ", round(Qtd_FEM. / Qtd_MASC., 1), ":1"),
      is.na(Qtd_FEM.) | Qtd_FEM. == 0 ~ "MASC 1:0",
      is.na(Qtd_MASC.) | Qtd_MASC. == 0 ~ "FEM 1:0",
      Qtd_MASC. == Qtd_FEM. ~ "IGUAL 1:1",
      TRUE ~ NA_character_
    )
  ) %>%
  select(Partido, Media_FEM., Media_MASC., Qtd_FEM., Qtd_MASC., Proporcao_Masc_Fem) %>%
  arrange(desc(coalesce(Media_MASC., 0) + coalesce(Media_FEM., 0)))

cat("Média da Receita, Quantidade de Eleitos e Proporção por Partido e Gênero:\n")
print(media_receita_corrigida, n = Inf, width = Inf)
```

```r
# Selecionar apenas duplicados com Cargo Prefeito ou Vice-prefeito
duplicados_prefeito_vice <- subset(duplicados_2020, Cargo %in% c("Prefeito", "Vice-prefeito"))

# Inicializar nova tabela com os duplicados atuais
complementados <- duplicados_prefeito_vice

# Loop por cada linha para buscar o par complementar (prefeito ou vice da mesma cidade/partido)
for (i in 1:nrow(duplicados_prefeito_vice)) {
  linha <- duplicados_prefeito_vice[i, ]
  cargo_atual <- linha$Cargo
  cidade <- linha$Cidade
  partido <- linha$Partido
  uf <- linha$UF
  
  # Se for Prefeito, buscar Vice-prefeito da mesma cidade, UF e partido
  if (cargo_atual == "Prefeito") {
    par <- subset(candidatos_2020,
                  Cidade == cidade &
                  UF == uf &
                  Cargo == "Vice-prefeito" &
                  Partido == partido)
  }
  # Se for Vice-prefeito, buscar Prefeito da mesma cidade, UF e partido
  else if (cargo_atual == "Vice-prefeito") {
    par <- subset(candidatos_2020,
                  Cidade == cidade &
                  UF == uf &
                  Cargo == "Prefeito" &
                  Partido == partido)
  }
  
  # Se encontrou par, adicionar às linhas complementadas
  if (nrow(par) > 0) {
    complementados <- rbind(complementados, par)
  }
}

# Remover possíveis duplicatas exatas
complementados <- unique(complementados[order(complementados$Cidade, complementados$Nome.Candidato), ])

# Exibir resultado
complementados
```
