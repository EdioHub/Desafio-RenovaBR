```python
import numpy as np
import matplotlib.pyplot as plt

def curva_lorenz(valores):
    """Retorna os pontos da curva de Lorenz para um array de valores positivos."""
    valores = np.sort(np.array(valores[valores > 0]))  # considera apenas valores > 0
    n = valores.size
    if n == 0:
        return np.array([0]), np.array([0])
    receita_acumulada = np.cumsum(valores)
    receita_acumulada = np.insert(receita_acumulada, 0, 0) / receita_acumulada[-1]
    proporcao_candidatos = np.linspace(0, 1, receita_acumulada.size)
    return proporcao_candidatos, receita_acumulada

# --- Plotando a curva de Lorenz para 'Votos' em cada ano ---
anos = df_votos['ano'].unique()
plt.figure(figsize=(8, 6))

for ano in sorted(anos):
    votos_ano = df_votos[df_votos['ano'] == ano]['Votos']
    x, y = curva_lorenz(votos_ano)
    plt.plot(x, y, label=f'{ano}')

# Linha de igualdade perfeita
plt.plot([0, 1], [0, 1], '--', color='gray', label='Igualdade perfeita')

plt.title('Curva de Lorenz – Distribuição de Votos por Ano')
plt.xlabel('Proporção acumulada de candidatos')
plt.ylabel('Proporção acumulada de votos')
plt.legend()
plt.grid(True)
plt.tight_layout()
# Se quiser salvar a figura, descomente a linha abaixo e ajuste o caminho
# plt.savefig('../figuras/p_curva_lorenz_votos.png', dpi=300)

plt.savefig('../figuras/p_lorenz_votos_2020_2024.png', dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def curva_lorenz(valores):
    """Retorna os pontos da curva de Lorenz"""
    valores = np.sort(np.array(valores[valores > 0]))
    n = valores.size
    if n == 0:
        return np.array([0]), np.array([0])
    receita_acumulada = np.cumsum(valores)
    receita_acumulada = np.insert(receita_acumulada, 0, 0) / receita_acumulada[-1]
    proporcao_candidatos = np.linspace(0, 1, receita_acumulada.size)
    return proporcao_candidatos, receita_acumulada

# Plotando a curva de Lorenz para cada ano
anos = df_filt['ano'].unique()
plt.figure(figsize=(8, 6))

for ano in sorted(anos):
    receitas = df_filt[df_filt['ano'] == ano]['Valor_Receita']
    x, y = curva_lorenz(receitas)
    plt.plot(x, y, label=f'{ano}')

# Linha de igualdade perfeita
plt.plot([0, 1], [0, 1], '--', color='gray', label='Igualdade perfeita')

plt.title('Curva de Lorenz - Receita de Campanha por Ano')
plt.xlabel('Proporção acumulada de candidatos')
plt.ylabel('Proporção acumulada da receita')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('../figuras/p_curva_lorenz_2020_2024.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Filtrar receitas positivas
df_plot = df_filt[df_filt['Valor_Receita'] > 0].copy()

# Padronizar colunas
df_plot['Genero'] = df_plot['Genero'].replace({'FEM.': 'FEMININO', 'MASC.': 'MASCULINO'})
df_plot['Raca'] = df_plot['Raca'].str.upper().str.strip()

# Cores por raça
cores_raca = {
    "PRETA": "#4F4F4F",
    "BRANCA": "#A9C0D8",
    "AMARELA": "#FFD700",
    "PARDA": "#8B4513",
    "INDÍGENA": "#F08080"
}

# Lista de raças que serão plotadas
racas_para_plotar = ["PRETA", "BRANCA", "PARDA", "AMARELA", "INDÍGENA"]

for raca in racas_para_plotar:
    df_raca = df_plot[df_plot['Raca'] == raca].copy()
    if df_raca.empty:
        continue

    # Criar coluna de grupo: Gênero - Ano
    df_raca['Grupo'] = df_raca['Genero'] + ' - ' + df_raca['ano'].astype(str)
    ordem_grupos = ['MASCULINO - 2020', 'FEMININO - 2020', 'MASCULINO - 2024', 'FEMININO - 2024']

    # Criar figura e eixos explicitamente
    fig, ax = plt.subplots(figsize=(8, 4))

    # Plot boxplot
    sns.boxplot(
        data=df_raca,
        x='Grupo',
        y='Valor_Receita',
        palette=[cores_raca[raca]] * 4,
        order=ordem_grupos,
        showfliers=False,
        ax=ax
    )

    # Plot stripplot
    sns.stripplot(
        data=df_raca,
        x='Grupo',
        y='Valor_Receita',
        color='black',
        order=ordem_grupos,
        alpha=0.3,
        jitter=True,
        size=2,
        ax=ax
    )

    ax.set_yscale('log')
    ax.set_xlabel('Gênero - Ano')
    ax.set_ylabel('Receita Total (escala log)')
    ax.set_title(f'Distribuição da Receita de Campanha - Raça: {raca.title()}')
    ax.grid(True, axis='y', linestyle='--', linewidth=0.5)
    fig.tight_layout()

    # Salvar antes de exibir
    nome_arquivo = f'../figuras/p_box_separado_receita_{raca.lower()}_genero_20_24.png'
    fig.savefig(nome_arquivo, dpi=300)

    # Exibir o plot
    plt.show()

    # Fechar figura para não manter em memória
    plt.close(fig)

print("Todos os gráficos foram exibidos e salvos em '../figuras/'.")
```

Ano de 2020<br>
Em 2020, observa-se uma leve superioridade na arrecadação média entre homens em comparação às mulheres, especialmente nas faixas superiores da distribuição. Em termos raciais, candidatos(as) brancos(as) de ambos os gêneros geralmente apresentaram maior arrecadação em percentis elevados. <br>
Mulheres negras (pretas e pardas) apresentaram menores receitas em quase todos os percentis, evidenciando uma dupla desigualdade por gênero e raça. A concentração de receita entre os 10% mais ricos foi especialmente acentuada entre homens brancos, com valores que superam meio milhão de reais.
<br>
Ano de 2024<br>
Em 2024, há avanços importantes na arrecadação feminina. Mulheres passaram a superar homens em alguns percentis intermediários (30º a 80º), especialmente entre candidatas pardas, pretas e indígenas. Este movimento sinaliza um fortalecimento da presença feminina nas campanhas com receitas medianas.
<br>
Apesar disso, os percentis mais altos (90º e 100º) ainda concentram majoritariamente os maiores valores entre os homens, especialmente brancos e pardos, o que indica que as campanhas com grande volume de recursos seguem sendo dominadas por eles.
<br>
Candidatos(as) indígenas também registraram aumentos expressivos, com medianas de receita significativamente superiores às de 2020, embora o grupo seja numericamente pequeno.
<br>
Destaques Gerais<br>
Houve um aumento expressivo na receita total em 2024 para todos os grupos, o que pode refletir mudanças nas estratégias partidárias, aumento do fundo eleitoral ou maior profissionalização das campanhas.
<br>
A escala logarítmica do gráfico evidencia a desigualdade: candidatos(as) no 100º percentil arrecadam centenas de vezes mais do que os(as) do 10º percentil.
<br>
Apesar de avanços femininos, sobretudo entre mulheres negras e indígenas, os homens brancos continuam no topo da distribuição de receitas.

Conclusão<br>
A análise da distribuição de receitas eleitorais por gênero e raça entre 2020 e 2024 revela uma realidade complexa: embora haja sinais claros de avanço na arrecadação entre mulheres — especialmente negras e indígenas —, as desigualdades estruturais permanecem significativas.
<br>
Mulheres, de forma geral, aumentaram sua presença nos percentis intermediários e passaram a disputar de forma mais equilibrada os recursos de campanha. Esse avanço é particularmente notável entre mulheres pretas, pardas e indígenas, que historicamente enfrentam múltiplas barreiras na política institucional. Os dados apontam que as campanhas de maior arrecadação continuam sendo um espaço restrito, refletindo desigualdades históricas de raça e gênero que seguem profundamente enraizadas.
<br>
Portanto, embora seja possível identificar progressos importantes, sobretudo no aumento da captação entre candidatas de grupos sub-representados, a equidade plena ainda está distante. A superação dessas disparidades requer não apenas a ampliação de candidaturas diversas, mas também a redistribuição efetiva dos recursos eleitorais de forma mais justa entre os diferentes grupos de gênero e raça.
