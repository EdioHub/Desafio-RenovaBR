Portanto, não basta apenas aumentar o número de candidaturas femininas — é fundamental garantir visibilidade, apoio político e equidade nas condições de disputa, para que a presença das mulheres na política se traduza efetivamente em representação.

```r
# 1. Contar o número de candidatos por partido
contagem_partidos <- as.data.frame(table(candidaturas_validadas$Partido))
colnames(contagem_partidos) <- c("Partido", "Candidatos")

# 2. Filtrar apenas os candidatos eleitos
candidatos_eleitos <- candidaturas_validadas %>%
  filter(Resultado_Eleitoral == "Eleito")

# 3. Contar o número de eleitos por partido
contagem_eleitos <- as.data.frame(table(candidatos_eleitos$Partido))
colnames(contagem_eleitos) <- c("Partido", "Eleitos")

# 4. Juntar as duas tabelas
tabela_produtividade <- merge(contagem_partidos, contagem_eleitos, by = "Partido", all.x = TRUE)

# 5. Substituir NAs por 0
tabela_produtividade$Eleitos[is.na(tabela_produtividade$Eleitos)] <- 0

# 6. Calcular produtividade (%)
tabela_produtividade$Produtividade <- round((tabela_produtividade$Eleitos / tabela_produtividade$Candidatos) * 100, 2)

# 7. Calcular quantidade de candidatos para eleger 1
tabela_produtividade$Candidatos_por_Eleito <- ifelse(
  tabela_produtividade$Eleitos > 0,
  round(tabela_produtividade$Candidatos / tabela_produtividade$Eleitos, 2),
  NA
)

# 8. Ordenar por produtividade decrescente
tabela_produtividade <- tabela_produtividade[order(-tabela_produtividade$Produtividade), ]

# 9. Exibir a tabela final
print(tabela_produtividade)

```

<span style="font-weight:bold; color:yellow; font-size:20px">
Número de Candidaturas Necessárias para Eleger 1 Vereador, por Raça</span>


```r
# Contagem de candidatos válidos por Raça no dataset 'candidaturas_validadas'
total_raca <- candidaturas_validadas %>%
  count(Raca, name = "Total")

# Contagem de eleitos por Raça no dataset 'candidatos_eleitos'
eleitos_raca <- candidatos_eleitos %>%
  count(Raca, name = "Eleitos")

# Juntar e calcular percentual de eleitos
comparativo_raca <- total_raca %>%
  left_join(eleitos_raca, by = "Raca") %>%
  mutate(
    Eleitos = ifelse(is.na(Eleitos), 0, Eleitos),            # Se não houver eleitos, coloca 0
    Percentual = round(Eleitos / Total * 100, 1),            # Percentual de eleitos
    Taxa_Sucesso = Eleitos / Total                           # Taxa de sucesso
  )

# Passo 1: Filtrar raças principais para equilíbrio (exclui AMARELA e INDÍGENA)
dados_principais <- comparativo_raca %>%
  filter(Raca %in% c("BRANCA", "PARDA", "PRETA"))

# Passo 2: Total de eleitos a ser redistribuído igualmente entre os 3 grupos
total_equilibrado <- sum(dados_principais$Eleitos)
eleitos_por_grupo <- total_equilibrado / 3

# Passo 3: Cálculo de candidaturas necessárias para equilíbrio proporcional
comparativo_raca <- comparativo_raca %>%
  mutate(
    Candidaturas_para_equilibrio = case_when(
      Raca %in% c("BRANCA", "PARDA", "PRETA") ~ round(eleitos_por_grupo / Taxa_Sucesso),
      TRUE ~ NA_real_
    )
  )

# Resultado final no console
cat("\nComparativo por Raça:\n")
print(comparativo_raca)

# Paleta de cores personalizada
cores_raca <- c(
  "PRETA" = "#4F4F4F",      # Cinza escuro
  "BRANCA" = "#A9C0D8",     # Azul acinzentado
  "AMARELA" = "#FFD700",    # Amarelo
  "PARDA" = "#8B4513",      # Marrom
  "INDÍGENA" = "#F08080"    # Vermelho claro
)

# Gráfico: Percentual de Eleitos por Raça (excluindo "SEM INFORMAÇÃO" apenas do gráfico)
ggplot(comparativo_raca %>% filter(Raca != "SEM INFORMAÇÃO"),
       aes(x = reorder(Raca, -Percentual), y = Percentual, fill = Raca)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = paste0(Percentual, "%")), vjust = -0.5, size = 5) +
  labs(
    title = "Percentual de Eleitos por Raça",
    y = "Percentual de Eleitos (%)",
    x = "Raça"
  ) +
  scale_fill_manual(values = cores_raca) +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 14, face = "bold"),
    axis.text = element_text(size = 12),
    legend.position = "none"
  )

```

Por outro lado, os candidatos pardos e, especialmente, pretos, enfrentam maiores obstáculos. Enquanto os pardos precisam de aproximadamente 30,9 candidaturas para cada eleito, os pretos enfrentam um desafio ainda maior: são necessárias 42,7 candidaturas para eleger um único vereador preto. Essa diferença é contundente, indicando que, proporcionalmente, os pretos precisam de 2,5 vezes mais candidaturas do que os brancos para obter o mesmo resultado. <br>
Se o objetivo fosse garantir igualdade no número de eleitos entre brancos e pretos, mantendo-se as taxas atuais de sucesso, seriam necessárias 6.448 candidaturas pretas — um número 4,2 vezes superior ao total de candidaturas pretas registradas. <br>

A eficiência eleitoral de brancos e pardos juntos é quase duas vezes maior do que a dos pretos. <br>

Esses números deixam claro que a sub-representação da população preta nas câmaras municipais não se resume à taxa de sucesso eleitoral. O maior entrave está na escassez de candidaturas viáveis, que limita estruturalmente a capacidade de representação política desse grupo. A equidade racial na política exige mais do que inclusão formal — requer ações afirmativas, estrutura de apoio e visibilidade real, de modo a permitir que essas candidaturas avancem com condições mínimas de disputa e competitividade em um sistema historicamente desigual. </span>
