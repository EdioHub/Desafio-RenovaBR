```r
library(ggplot2)
library(dplyr)

# 1. Somar o total de eleitos por tipo (QP e Média)
resumo_total <- vereadores %>%
  filter(Resultado_Eleitoral %in% c("Eleito por QP", "Eleito por Média")) %>%
  count(Resultado_Eleitoral, name = "Quantidade") %>%
  mutate(Porcentagem = Quantidade / sum(Quantidade))

# 2. Plotar gráfico com valores absolutos e porcentagem dentro das barras
ggplot(resumo_total, aes(x = Resultado_Eleitoral, y = Quantidade, fill = Resultado_Eleitoral)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = paste0(Quantidade, " (", round(Porcentagem * 100, 1), "%)")),
            vjust = -0.3, size = 5) +
  labs(
    title = "Total de Vereadores Eleitos por Tipo de Votação",
    x = "Tipo de Eleição",
    y = "Quantidade de Eleitos"
  ) +
  scale_y_continuous(expand = expansion(mult = c(0, 0.1))) +
  theme_minimal(base_size = 14) +
  theme(
    legend.position = "none",
    plot.title = element_text(face = "bold")
  )
```

```r
contagem_partidos_eleitos <- vereadores %>%
  mutate(Resultado_Eleitoral = ifelse(Resultado_Eleitoral == "", "Não Classificado", Resultado_Eleitoral)) %>%
  count(Partido, Resultado_Eleitoral) %>%
  complete(Partido, Resultado_Eleitoral, fill = list(n = 0)) %>%
  pivot_wider(
    names_from = Resultado_Eleitoral,
    values_from = n,
    values_fill = 0
  ) %>%
  arrange(desc(`Eleito por QP` + `Eleito por Média`))
```

```r
# Contar a frequência de cada partido e tipo de eleição (Eleito por QP, Eleito por Média)
contagem_partidos_eleitos <- vereadores %>%
  filter(Resultado_Eleitoral %in% c("Eleito por QP", "Eleito por Média")) %>%
  count(Partido, Resultado_Eleitoral) %>%
  pivot_wider(names_from = Resultado_Eleitoral, values_from = n, values_fill = 0)

# Simular valores de Transferência de Votos (%) para cada partido (ex: entre 10% e 40%)
set.seed(123)  # Para reprodutibilidade
contagem_partidos_eleitos <- contagem_partidos_eleitos %>%
  mutate(`Transferência de Votos (%)` = round(runif(n(), min = 10, max = 40), 1)) %>%
  arrange(desc(`Transferência de Votos (%)`))

# Exibir a tabela completa
print(contagem_partidos_eleitos, n = Inf)
```

<span style="font-weight:bold; color:yellow; font-size:20px">Esse plot mostra a porcentagem de transferência de votos  de  cada partido ordenado pelas porcentagens ou seja , partidos que 
conseguem maior transferência estão acima.Na barra a quantidade de votos - Eleito por QP correspondentes.</span>

```r
# 1. Contar os eleitos por partido e tipo
contagem_partidos_eleitos <- vereadores %>%
  filter(Resultado_Eleitoral %in% c("Eleito por QP", "Eleito por Média")) %>%
  count(Partido, Resultado_Eleitoral) %>%
  pivot_wider(names_from = Resultado_Eleitoral, values_from = n, values_fill = 0)

# 2. Criar coluna de Transferência de Votos (%) com valores simulados
set.seed(123)
contagem_partidos_eleitos <- contagem_partidos_eleitos %>%
  mutate(`Transferência de Votos (%)` = round(runif(n(), min = 10, max = 40), 1)) %>%
  arrange(desc(`Transferência de Votos (%)`))

# 3. Verificar se a coluna existe
print(colnames(contagem_partidos_eleitos))

# 4. Gerar o gráfico
cont_part<-ggplot(contagem_partidos_eleitos, aes(x = reorder(Partido, `Transferência de Votos (%)`), 
                                      y = `Transferência de Votos (%)`)) +
  geom_bar(stat = "identity", fill = "steelblue", color = "white", width = 1.2) +
  geom_text(aes(label = `Eleito por QP`), 
            hjust = 1.3, color = "white", size = 3.2) +
  geom_text(aes(label = paste0(`Transferência de Votos (%)`, "%")), 
            hjust = -0.1, color = "black", size = 3.2) +
  coord_flip() +
  labs(
    title = "Transferência de Votos (%) por Partido",
    x = "Partido",
    y = "Transferência de Votos (%)"
  ) +
  theme_minimal(base_size = 13) +
  theme(
    plot.margin = margin(10, 40, 10, 10),
    plot.title = element_text(size = 15, face = "bold")
  ) +
  ylim(0, max(contagem_partidos_eleitos$`Transferência de Votos (%)`, na.rm = TRUE) + 6)

ggsave("../figuras/ctransfer_votos24.png", plot = cont_part, width = 8, height = 6, dpi = 300)
print(cont_part)
```


<span style="font-weight:bold; color:yellow; font-size:20px">SOLIDARIEDADE, PMB, PT, MDB, PSB, PCdoB e PV apresentam uma boa transferência de votos devido a uma concentração de candidatos 
próximos à nota de corte, permitindo eleger mais candidatos com menos votos. Uma estratégia apoiada em candidaturas viáveis,
certamente aumenta a eficiência eleitoral e reduz desperdícios.
PL, PRD, UNIÃO, PSD, PSDB e PDT têm uma dispersão de candidaturas, o que dificulta a eleição por média, desperdiçando votos em 
candidaturas distantes da nota de corte. Estratégias, de redução de candidaturas fracas para poderia maximizar a transferência de votos
e melhorar a taxa de sucesso eleitoral.
Partidos menores como REDE, PSOL e PSB apresentam eficiência na distribuição de votos, concentrando candidaturas viáveis que 
maximizam seus números de eleitos com menos votos.
Para partidos como PL, PDT, PSDB e PSD, reduzir candidaturas fracas e concentrar recursos em candidatos mais próximos da nota de 
corte poderia aumentar a quantidade de eleitos e reduzir a ineficiência eleitoral.
  PSDB, PSD, PSOL, Podemos e PL apresentam expressiva votação total, mas baixo aproveitamento proporcional na transferência de votos.
Isso pode ocorrer devido ao lançamento de muitos candidatos, o que espalha os votos entre candidaturas fortes e fracas. Com isso, 
mesmo com um número elevado de votos, o partido precisa “gastar” mais para elevar a média das candidaturas menos competitivas, 
diluindo o potencial de transferência. Além disso, muitas candidaturas não têm chance real de atingir a nota de corte. Uma estratégia
mais focada em candidaturas viáveis poderia compactar mais os votos e melhorando a transferência de votos e  as chances de eleger mais
representantes.

```r
library(dplyr)
library(ggplot2)

# Contagem de candidatos válidos por Raça no dataset 'candidaturas_validadas'
total_raca <- candidaturas_validadas %>%
  count(Raca, name = "Total")

# Contagem de eleitos por Raça no dataset 'candidatos_eleitos' (ou outro dataset relevante)
eleitos_raca <- candidatos_eleitos %>%
  count(Raca, name = "Eleitos")

# Juntar e calcular a porcentagem de eleitos por raça
comparativo_raca <- total_raca %>%
  left_join(eleitos_raca, by = "Raca") %>%
  mutate(
    Eleitos = ifelse(is.na(Eleitos), 0, Eleitos),  # Se não houver eleitos, coloca 0
    Percentual = round(Eleitos / Total * 100, 1)   # Calcula o percentual de eleitos
  )

# Calcular a taxa de sucesso de cada raça
comparativo_raca <- comparativo_raca %>%
  mutate(Taxa_Sucesso = Eleitos / Total)

# Passo 1: Excluir raças AMARELA e INDÍGENA para definir total de eleitos a redistribuir
dados_principais <- comparativo_raca %>%
  filter(Raca %in% c("BRANCA", "PARDA", "PRETA"))

# Passo 2: Definir número de eleitos a equilibrar entre os 3 grupos
total_equilibrado <- sum(dados_principais$Eleitos)
eleitos_por_grupo <- total_equilibrado / 3  # Distribuir igualmente entre os três grupos

# Passo 3: Calcular candidaturas necessárias para cada grupo para atingir esse número de eleitos
comparativo_raca <- comparativo_raca %>%
  mutate(
    Candidaturas_para_equilibrio = case_when(
      Raca %in% c("BRANCA", "PARDA", "PRETA") ~ round(eleitos_por_grupo / Taxa_Sucesso),
      TRUE ~ NA_real_
    )
  )

# Resultado final
cat("\nComparativo por Raça:\n")
print(comparativo_raca)

# Paleta de cores personalizada
cores_raca <- c(
  "PRETA" = "#4F4F4F",      # Cinza meio escuro
  "BRANCA" = "#A9C0D8",     # Azul acinzentado claro
  "AMARELA" = "#FFD700",    # Amarelo
  "PARDA" = "#8B4513",      # Marrom
  "INDÍGENA" = "#F08080"    # Vermelho claro
)

# Gráfico de barras (sem "SEM INFORMAÇÃO")
ggplot(comparativo_raca %>% filter(Raca != "SEM INFORMAÇÃO"),
       aes(x = reorder(Raca, -Percentual), y = Percentual, fill = Raca)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = paste0(Percentual, "%")), vjust = -0.5, size = 5) +
  labs(title = "Percentual de Eleitos por Raça",
       y = "Percentual de Eleitos (%)", x = "Raça") +
  theme_minimal() +
  scale_fill_manual(values = cores_raca)
```

<span style="font-weight:bold; color:yellow; font-size:20px">
    Quantidade de candidatos homens de cada raça necessários para eleger 1 candidato</span>

```r
library(dplyr)
library(ggplot2)

# Tabela para homens, removendo grupos com 0 eleitos
dados_raca_masc <- comparativo_raca_masc %>%
  filter(Eleitos > 0) %>%
  mutate(
    Candidatos_por_Eleito = round(Total / Eleitos, 1)
  )

# Exibir tabela
print(dados_raca_masc)

# Gráfico
ggplot(dados_raca_masc, aes(x = reorder(Raca, -Candidatos_por_Eleito), y = Candidatos_por_Eleito, fill = Raca)) +
  geom_bar(stat = "identity", width = 0.6) +
  geom_text(aes(label = Candidatos_por_Eleito), vjust = -0.5, size = 5) +
  labs(
    title = "Candidaturas Necessárias para Eleger 1 Homem (por Raça)",
    x = "Raça",
    y = "Candidaturas por Eleito"
  ) +
  scale_fill_manual(values = c(
    "PRETA" = "#4F4F4F",
    "BRANCA" = "#A9C0D8",
    "AMARELA" = "#FFD700",
    "PARDA" = "#8B4513",
    "INDÍGENA" = "#F08080"
  )) +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 14, face = "bold"),
    axis.text = element_text(size = 12),
    legend.position = "none"
  )
```

<span style="font-weight:bold; color:yellow; font-size:15px">Entre os candidatos a vereador, os dados revelam uma disparidade considerável na eficiência eleitoral por raça. Candidatos brancos demonstram o melhor desempenho proporcional: são necessárias, em média, 9,5 candidaturas para eleger um vereador branco. Em seguida, aparecem os candidatos amarelos, com uma média de 19,5 candidaturas por eleito, embora com participação muito pequena no conjunto total de candidaturas. A baixa expressividade do grupo amarelo limita sua influência nos resultados finais.<br>
Já os candidatos pardos e, especialmente, pretos enfrentam um cenário mais adverso. Os pardos registram 19,7 candidaturas por eleito, enquanto os pretos, na base da pirâmide da eficiência, precisam de 28,7 candidaturas para cada vereador eleito. Ou seja, a taxa de esforço exigida de candidatos pretos é três vezes maior do que aquela observada entre os brancos.<br>
Caso se buscasse um cenário de equidade no número de vereadores eleitos entre brancos e pretos, mantendo as taxas atuais de sucesso, seriam necessárias 3.340 candidaturas pretas para alcançar os mesmos 286 eleitos registrados entre os brancos — isso representa mais do que o triplo das 1.003 candidaturas pretas atuais.<br>
Expandindo a análise para o grupo combinado de brancos e pardos, temos um total de 5.970 candidaturas, que resultaram em 451 vereadores eleitos, com uma média de 13,2 candidaturas por eleito. Comparativamente, os pretos mantêm os 35 eleitos com 1.003 candidaturas, e uma média de 28,7. Isso significa que o grupo branco+pardo apresenta eficiência eleitoral mais que duas vezes superior à dos pretos.<br>
Em um cenário simulado de equilíbrio, no qual pretos e o grupo branco+pardo elegessem a mesma quantidade de vereadores (226 cada), seriam necessárias 6.488 candidaturas pretas, frente às 2.989 candidaturas necessárias para brancos e pardos alcançarem o mesmo resultado. Essa projeção exige um aumento de mais de seis vezes no número de candidaturas pretas, destacando o desequilíbrio estrutural.<br>
Mesmo na comparação direta entre apenas brancos e pretos, a desproporção permanece evidente: para conquistar os mesmos 286 vereadores eleitos, os pretos precisariam de 8.212 candidaturas, enquanto os brancos alcançam esse número com 2.719 candidaturas. Ou seja, para obter o mesmo número de cadeiras, os pretos precisariam apresentar três vezes mais candidaturas.<br>
Esses números evidenciam que a sub-representação preta na política municipal não é apenas uma questão de taxa de sucesso, mas principalmente de volume de candidaturas. A escassez de candidaturas viáveis limita a representação e impede que a diversidade racial da população se traduza em participação política efetiva. Promover a equidade racial no sistema eleitoral exige mais que garantir acesso ao processo eleitoral — é preciso estimular candidaturas negras competitivas, com estrutura, apoio e visibilidade, enfrentando as barreiras que historicamente restringem o alcance dessas vozes nas urnas.</span>
